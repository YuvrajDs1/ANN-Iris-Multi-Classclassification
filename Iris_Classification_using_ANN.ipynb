```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df
```

```python
iris_df['target'] = iris['target']
iris_df
```

```python
# in dataset -> 0 = Setosa, 1 = Versicolor, 2 = Virginica
iris_df['target'].value_counts()
```

```python
iris_df.info()
```

```python
iris_df.describe()
```

### Since all data is numeric no need for preprocessing

```python
# Split data
X = iris_df.drop('target', axis=1)
y = iris_df['target']
X,y
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

## ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
(X_train.shape[1],)
```

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(3, activation='softmax')
])
```

```python
model.summary()
```

```python
# Compile our model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Early Stopping
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
from tensorflow.keras.callbacks import TensorBoard
import datetime
```

```python
log_dir = 'logs/fit/' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
model.fit(
    X_train, y_train, epochs=50, validation_data=(X_test, y_test), callbacks=[early_stopping_callback, tensorboard_callback]
)
```

```python
model.save('iris_classifier.h5')
```
